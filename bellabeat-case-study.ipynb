```python
# Data Cleaning

df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format= '%m/%d/%Y')
df['Id'] = df['Id'].astype(str)
df.dtypes
```

```python
df['distance_diff'] = df['TotalDistance'] - df['TrackerDistance']
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.rename(columns = {'activitydate':'activity_date', 'totalsteps':'total_steps', 'totaldistance':'total_distance', 
                     'trackerdistance':'tracker_distance','loggedactivitiesdistance':'logged_activities_distance', 
                     'veryactivedistance':'very_active_distance', 'moderatelyactivedistance':'moderately_acvtive_distance', 
                     'lightactivedistance':'light_active_distance', 'sedentaryactivedistance':'sedentary_active_distance', 
                     'veryactiveminutes':'very_active_minutes', 'fairlyactiveminutes':'fairly_active_minutes',
                     'lightlyactiveminutes':'lightly_active_minutes', 'sedentaryminutes':'sedentary_minutes'}, inplace=True)
df.columns
```

```python
# Creating New Columns
df['active_time'] = df['very_active_minutes'] + df['fairly_active_minutes'] + df['lightly_active_minutes']

df["active_time"] = df["active_time"].astype(int)
```

```python
df['total_time'] = df['very_active_minutes'] + df['fairly_active_minutes']+ df['lightly_active_minutes'] + df['sedentary_minutes']

df["total_time"] = df["total_time"].astype(int)
```

```python
day_of_week = df['activity_date'].dt.day_name()
df['day_of_week'] = day_of_week

df['n_day_of_week'] = df['activity_date'].dt.weekday #Days are in numerical order starting with Monday and 0
```

```python
# Subsetting The Data
df = df[['id', 'activity_date', 'total_steps', 'total_distance',
       'very_active_minutes', 'fairly_active_minutes',
       'lightly_active_minutes', 'sedentary_minutes', 'calories',
    'day_of_week', 'n_day_of_week','total_time','active_time',]].copy()
```

```python
# 1440 min is total amount of min in one day

df.sort_values(by=['total_time'], ascending=False)
```

```python
# Sedentary: less than 7500 steps on average
# Active: between 7500 and 12500 steps on average
# Very Active: more than 12500 steps on average

id_grp = df.groupby(['id'])
id_avg_step = id_grp['total_steps'].mean().sort_values(ascending=False)
id_avg_step = id_avg_step.to_frame()

id_avg_step

conditions = [
    (id_avg_step <= 7500),
    (id_avg_step > 7500) & (id_avg_step < 12500),
    (id_avg_step >= 12500)
]
values = ['Sedentarty','Active','Very Active']

id_avg_step['intensity'] = np.select(conditions,values)

id_activity_level = id_avg_step['intensity']

df['intensity'] = [id_activity_level[c] for c in df['id']]
```

```python
# Low: less than 360 minutes
# Med: between 360 and 720 minutes
# High: more than 720 minutes
id_grp = df.groupby(['id'])

conditions = [
    (df['total_time'] <= 360),
    (df['total_time'] > 360) & (df['total_time'] < 720),
    (df['total_time'] >= 720)
]
values = ['Low','Med','High']

df['usage'] = np.select(conditions,values)
```

```python
# Your 'calories' and 'total_time' data
calories_data = df['calories']
total_time_data = df['total_time']

# Scale the 'calories' for better visualization
scaler = MinMaxScaler()
calories_scaled = scaler.fit_transform(calories_data.values.reshape(-1, 1)).flatten()

# Create a scatter plot with color representing 'calories' and smaller marker size
plt.scatter(total_time_data, calories_data, c=calories_scaled, cmap='viridis_r', s=50, alpha=0.8)

# Customize the plot
plt.xlabel('Total Time (minutes)')
plt.ylabel('Calories')
plt.title('Scatter Plot: Total Time vs Calories')
plt.colorbar(label='Scaled Calories')  # Add colorbar to show the scale of 'calories'

# Display the plot
plt.show()
```

```python
#steps per day on average
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

day_grp = df.groupby(['day_of_week'])
avg_daily_time = day_grp['active_time'].mean()
avg_time = df['active_time'].mean()

# Use consistent coloring for the bars
bars = plt.bar(avg_daily_time.index, avg_daily_time, color='skyblue', label='Avg Daily Time')

days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

ax.set_xticks(range(len(days_of_week)))
ax.set_xticklabels(days_of_week)

# Use a dashed line for the average daily time reference line
ax.axhline(y=avg_daily_time.mean(), color='blue', linestyle='--', label='Avg Daily Time')

# Add data labels to each bar with better centering
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2,
             f'{bar.get_height():.2f}', ha='center', va='bottom', color='black')

ax.set_xlabel('Day of week')
ax.set_ylabel('Average Time (minutes)')
ax.set_title('Average Active Time per Day of the Week')

# Move the legend outside the plot
ax.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Calculating the total minutes for each activity level
very_active_mins = df['very_active_minutes'].sum()
fairly_active_mins = df['fairly_active_minutes'].sum()
lightly_active_mins = df['lightly_active_minutes'].sum()
sedentary_mins = df['sedentary_minutes'].sum()

# Creating data slices for the pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]

# Defining labels for each activity level
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']

# Configuring the extent of 'explode' for visual emphasis
explode = [0.5, 0.5, 0.2, 0]

# Specifying colors for each activity level
colors = ['tab:blue', 'tab:orange', 'tab:gray', 'tab:cyan']

# Creating a pie chart with the specified parameters
plt.pie(slices, labels=labels, explode=explode, colors=colors, autopct='%1.1f%%')

# Adding a title to the pie chart
plt.title('Intensity (In Minutes)')

# Displaying the pie chart
plt.show()
```
